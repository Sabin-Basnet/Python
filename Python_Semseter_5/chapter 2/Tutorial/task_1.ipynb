```python
# TASK 
# 1> Create a table for Books with fields : title, author , year
# 2> Insert 3 books and query by author
# 3> Update a book's year
# 4> Delete a book
```

```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import declarative_base, sessionmaker

#create engine
engine= create_engine('sqlite:///books.db', echo=False)

Base=declarative_base()
# ORM --> Object Relational Mapping
# ORM maps classes → tables and objects → rows
```

```python
# creates books table
class Book(Base):
    __tablename__ = 'books'
    id= Column(Integer, primary_key=True)
    title=Column(String)
    author=Column(String)
    year=Column(String)

# create table in database
Base.metadata.create_all(engine)

#create Session
Session = sessionmaker(bind=engine)
session = Session()
```

```python
# insert data

book1 = Book(title="Math", author="Ram", year="2000")
book2 = Book(title="Science", author="Hari", year="2003")
book3 = Book(title="Computer", author="Shyam", year="2004")

# add data to database
session.add(book1)
session.add_all([book2, book3])

session.commit()
```

```python
# query using author
print("Books by Ram:")
books_by_ram = session.query(Book).filter_by(author="Ram").all()
for book in books_by_ram:
    print(book.title, book.year)
```

```python
# update book year 
book_to_update = session.query(Book).filter_by(title="Science").first()    #use first() to select first row with title="Science"
book_to_update.year = 2010
session.commit()
print("\nAfter updating year of 'Science':")
updated_book = session.query(Book).filter_by(title="Science").first()
print(updated_book.title, updated_book.year)
```

```python
book_to_delete = session.query(Book).filter_by(title="Computer").first()

if book_to_delete:
    session.delete(book_to_delete)
    session.commit()

print("Books after deletion:")
all_books = session.query(Book).all()
for book in all_books:
    print(book.title, book.author, book.year)
```

```python
session.close()
```
