```python
# Task: Excel File Manipulation
# Objective: Read an Excel file, process the data, and write the results to a new Excel file.
# Instructions:
# You are provided with an Excel file named students.xlsx containing the following columns:
# Name
# Math
# Science
# English
# Write a Python program that:
# a. Reads the Excel file into a DataFrame (or using openpyxl if preferred).
# b. Calculates the Total and Average marks for each student.
# c. Adds these as new columns in the DataFrame or Excel sheet.
# d. Identifies students who scored average >= 60 as "Passed" and < 60 as "Failed". Add a column Result for this.
# e. Writes the updated data to a new Excel file students_result.xlsx.
```

```python
import pandas as pd

# read excel file into a dataframe
df=pd.read_excel("students.xlsx")
print("Original data:")
print(df)
```

```python
# calculate total and average marks for each student and add them as new columns

df['Total'] = df['Math'] + df['Science'] + df['English']
df['Average'] = df['Total'] / 3
```

```python
df
```

```python
# if average > 60 then passs, else fail

df['Result'] = df['Average'].apply(lambda x: 'Passed' if x >= 60 else 'Failed')
```

```python
df
```

```python
# write the updated data into new excel file

df.to_excel("students_result.xlsx", index=False)

print("Data Updated and Saved into new excel file")
```
