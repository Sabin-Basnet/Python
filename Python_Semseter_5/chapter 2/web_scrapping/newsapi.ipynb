```python
import requests
from bs4 import BeautifulSoup
from flask import Flask, jsonify
```

```python

app = Flask(__name__)

def scrape_kantipur():
    url = "https://ekantipur.com/latest-news"
    # Adding a Header makes you look like a real browser so they don't block you
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    articles = []
    
    # Kantipur usually places headlines inside <article> tags or <h2> within a 'teaser' class
    for item in soup.select('article h2 a'):
        title = item.get_text(strip=True)
        link = item['href']
        
        # Ensure the link is a full URL
        if link.startswith('/'):
            link = f"https://ekantipur.com{link}"
            
        articles.append({
            "title": title,
            "link": link
        })
        
        # Limit to top 15 news items
        if len(articles) >= 15:
            break
            
    return articles

@app.route('/get-kantipur-news', methods=['GET'])
def get_news_api():
    try:
        data = scrape_kantipur()
        return jsonify({
            "status": "success",
            "source": "Kantipur News",
            "news": data
        })
    except Exception as e:
        return jsonify({"status": "error", "message": str(e)})

if __name__ == '__main__':
    app.run(debug=True)
```
