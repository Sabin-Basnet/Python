```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://wikimedia.org'
response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    # print(soup.prettify())
    data = soup.find_all('h2')
    for i in data:
        print(i.text)
else:
    print("Response = ",requests.status_codes)
```

```python
news=[]
for tag in soup.find_all(['h2','h3']):
    a = tag.find('a')
    if a and a.get("href"):
        title=  a.get_text(strip=True)
        link= a['href']
        if title:
            news.append({
                'headline': title,
                'link': link
        })
    
df = pd.DataFrame(news)
df.drop_duplicates(subset='headline').reset_index(drop = True)
df.to_csv('links_2.csv', index = False)
```
