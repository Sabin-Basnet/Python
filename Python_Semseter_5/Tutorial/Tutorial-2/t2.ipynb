```python
#  1. Write an iterator that returns the first n even numbers.
class EvenNumbers:
    def __init__(self, n):
        self.n = n
        self.count = 0
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            num = self.current
            self.current += 2
            self.count += 1
            return num
        raise StopIteration
    

n=int(input("Enter 'n':"))

numbers = EvenNumbers(n)
for num in numbers:
    print(num)
```

```python
# 2. Write a generator that yields the Fibonacci sequence up to n terms.

def fibonacci_gen(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        # Logic: Update a to the next number, and b to the sum of the two
        a, b = b, a + b

# Usage
n = int(input("Enter number of terms: "))
fib_sequence = fibonacci_gen(n)

for num in fib_sequence:
    print(num)

```

```python
# 3. Implement a decorator that prints “Start” before a function and “End” after it.
def my_decorator(func):
    def wrapper():
        print("Start")      # Logic before the function
        func()              # The actual function runs
        print("End")        # Logic after the function
    return wrapper

# Using the decorator with the @ symbol
@my_decorator
def say_hello():
    print("Hello!")

# Execution
say_hello()
```

```python
# 5. Implement an iterator class that iterates over a range in reverse order.
class ReverseRange:
    def __init__(self, start, end):
        # We start at the end (minus 1 to stay within bounds)
        self.current = end -1
        self.start = start

    def __iter__(self):
        return self

    def __next__(self):
        # As long as we haven't gone below the start
        if self.current >= self.start:
            num = self.current
            self.current -= 1  # Move backward
            return num
        # Stop when the countdown is finished
        raise StopIteration

# Usage
rev = ReverseRange(1, 5)
for n in rev:
    print(n)
```

```python
# 7. Create a decorator to measure memory usage of a function.
import os
import psutil

def monitor_memory(func):
    def wrapper(*args, **kwargs):
        # 1. Check memory before the function runs
        process = psutil.Process(os.getpid())
        mem_before = process.memory_info().rss
        
        # 2. Run the actual function
        result = func(*args, **kwargs)
        
        # 3. Check memory after the function runs
        mem_after = process.memory_info().rss
        
        # 4. Calculate the difference
        print(f"Memory used by '{func.__name__}': {mem_after - mem_before} bytes")
        return result
    return wrapper

@monitor_memory
def create_large_list():
    return [i for i in range(1000000)]

# Execution
create_large_list()
```

```python
# 8. Write a generator that yields prime numbers indefinitely.
def infinite_primes():
    num = 2
    while True:
        # A number is prime if it isn't divisible by any number 
        # between 2 and its own square root
        if all(num % i != 0 for i in range(2, int(num**0.5) + 1)):
            yield num
        num += 1

# Usage
primes = infinite_primes()

# Get the first 5 primes
for _ in range(5):
    print(next(primes))
```

```python
# 9. Create a decorator to cache results of a function (memoization).
def memoize(func):
    # This dictionary acts as the function's memory
    cache = {}
    
    def wrapper(*args):
        # If we have seen these arguments before, return the saved result
        if args in cache:
            print(f"Fetching from cache for: {args}")
            return cache[args]
        
        # Otherwise, calculate it and save it in the cache
        print(f"Calculating result for: {args}")
        result = func(*args)
        cache[args] = result
        return result
        
    return wrapper

@memoize
def expensive_calculation(a, b):
    return a + b

# Execution
print(expensive_calculation(10, 5)) # Calculates
print(expensive_calculation(10, 5)) # Fetches from cache (instant!)
```

```python
# 10. Implement a generator that filters only even numbers from a list.
def filter_evens(numbers):
    for num in numbers:
        if num % 2 == 0:
            yield num

# Usage
my_list = [1, 5, 8, 12, 13, 18, 21]
even_gen = filter_evens(my_list)

# Printing the results
for even in even_gen:
    print(even)
```

```python
# 11. Write a generator to simulate a stream of sensor readings.
import random
import time

def sensor_stream(limit=10):
    count = 0
    while count < limit:
        # Simulate a temperature reading between 20.0 and 30.0
        reading = round(random.uniform(20.0, 30.0), 2)
        yield f"Temperature: {reading}°C"
        
        # Simulate a delay between readings (0.5 seconds)
        time.sleep(0.5)
        count += 1

# Usage
for data in sensor_stream(5):
    print(data)
```

```python
# 12. Implement a decorator to log execution time and input arguments.
import time

def log_details(func):
    def wrapper(*args, **kwargs):
        # 1. Record the start time
        start_time = time.perf_counter()
        
        # 2. Run the actual function
        result = func(*args, **kwargs)
        
        # 3. Record the end time
        end_time = time.perf_counter()
        
        # 4. Log the details
        duration = end_time - start_time
        print(f"--- Log Entry ---")
        print(f"Function: {func.__name__}")
        print(f"Arguments: {args}")
        print(f"Execution Time: {duration:.6f} seconds")
        print(f"-----------------")
        
        return result
    return wrapper

@log_details
def add_numbers(a, b):
    time.sleep(0.1) # Simulate a slight delay
    return a + b

# Execution
add_numbers(10, 20)
```

```python
# 13. Write an iterator that yields only unique items from a list.
class UniqueIterator:
    def __init__(self, data):
        self.data = data
        self.seen = set()  # This acts as the iterator's memory
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        while self.index < len(self.data):
            item = self.data[self.index]
            self.index += 1
            
            # If we haven't seen this item before, record it and return it
            if item not in self.seen:
                self.seen.add(item)
                return item
        
        # If the loop finishes, we've checked everything
        raise StopIteration

# Usage
my_list = [10, 20, 10, 30, 20, 40]
unique_items = UniqueIterator(my_list)

for item in unique_items:
    print(item)
```

```python
# 14. Implement a decorator to check if inputs of a function are positive numbers.
def ensure_positive(func):
    def wrapper(*args):
        # Check every argument passed to the function
        for arg in args:
            if arg <= 0:
                # If a negative number or zero is found, stop everything
                raise ValueError(f"Invalid input: {arg}. Only positive numbers allowed!")
        
        # If all checks pass, run the function
        return func(*args)
    return wrapper

@ensure_positive
def calculate_area(length, width):
    return length * width

# Execution
print(calculate_area(5, 10))  # Works fine
# print(calculate_area(5, -2)) # This would trigger the ValueError
```

```python
# 15. Write a generator to simulate an infinite sequence of random numbers.

import random

def infinite_random_gen(low, high):
    while True:
        # Generate a random integer between low and high (inclusive)
        yield random.randint(low, high)

# Usage
random_stream = infinite_random_gen(1, 100)

# Pulling 5 random numbers from the infinite stream
print("First 5 random numbers:")
for _ in range(5):
    print(next(random_stream))
```
