```python
def fibo(n):
    a,b =0,1
    for i in range(n):
        yield a 
        a,b = b,a+b

f = fibo(15)
for num in f:
    print(num)
```

```python
def deco(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper
@deco
def demo():
    print("This is a function.")

demo()
```

```python
import csv

def csv_reader(file_path):
    with open(file_path, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            yield row
```
