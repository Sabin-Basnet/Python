```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5] # e.g., Study Hours
y = [10, 25, 35, 50, 60] # e.g., Exam Scores

# Plotting
plt.scatter(x, y, color='blue')

# Labeling (Important for full marks!)
plt.title("Scatter Plot Example")
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")

plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Create a figure with 1 row and 2 columns
fig, ax = plt.subplots(1 ,2, figsize=(10, 4))

# 2. First plot (Index 0)
ax[0].plot([1, 2, 3], [10, 20, 30], color='red')
ax[0].set_title("Line Plot")

# 3. Second plot (Index 1)
ax[1].scatter([1, 2, 3], [10, 20, 30], color='blue')
ax[1].set_title("Scatter Plot")

# 4. Automatically adjust spacing
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Create a sample correlation matrix
data = {
    'Price': [100, 120, 110, 150],
    'Volume': [50, 45, 60, 80],
    'Profit': [20, 25, 22, 35]
}
df = pd.DataFrame(data)
corr_matrix = df.corr() # Calculates correlation between -1 and 1

# 2. Plotting the Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```

```python
import schedule
import time
from datetime import datetime

def write_file():
    with open("schedule_demo.txt",'a') as f:
        f.write(f"time: {datetime.now()}\n")
    
schedule.every(1).minutes.do(write_file)
while(True):
    schedule.run_pending()
    time.sleep(1)
```
