```python
# merging
import pandas as pd
df_tab1 = pd.DataFrame({
    'id': [1,2,3,4],
    'name': ['ram', 'shyam', 'hari','']
})

df_tab2 = pd.DataFrame({
    'id': [1,2,3,4,5],
    'marks':[20,30, 55, 43, 22]
})

df_tab3= pd.merge(df_tab1, df_tab2 , on='id', how='outer')
df_tab3
```

```python
# joining--> combined df using index instead of columns
df_tab1 = pd.DataFrame({
    # 'id': [1,2,3,4],
    'name': ['ram', 'shyam', 'hari','unknown']
}, index=[1,2,3,4])

df_tab2 = pd.DataFrame({
    # 'id': [1,2,3,4,5],
    'marks':[20,30, 55, 43, 22]
}, index=[1,2,3,4,5])

df_tab3= df_tab1.join(df_tab2, how='inner')
df_tab3
# its use case is little different, use it for joining different
#  measurements like temp in table 1 and humidity in table 2
```

```python
# melt -- wide --> long
df_scores = pd.DataFrame({
    'Student':['Ram', 'Hari'],
    'math': [30, 19],
    'english': [12, 32]
})
print(df_scores) # this format is wide format

# lets change it into long format
melted = pd.melt(
    df_scores,
    id_vars=['Student'],
    var_name='Subject',
    value_name='Score'
)
melted
```

```python
# pivot -- long-->wide
df_long = pd.DataFrame({
    'Student':['Ram', 'Ram','Hari', 'Hari'],
    'Subject': ['math', 'english','math', 'english'],
    'Score': [10, 43, 54, 23]
})
print(df_long)

pivoted = df_long.pivot(
    index='Student',
    columns='Subject',
    values='Score'
)
pivoted
```

```python
# another method is stack
stacked = df_long.stack()
print(stacked)
```

```python
from sklearn.preprocessing import MinMaxScaler
x = np.array([[200],[300],[400],[600],[1000]])
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(x)
print(scaled_data)
# range(0,1)
```

```python
# z-score normalization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(x)
scaled_data
```

```python
# label encoding -->best to use for ordinal data(not nomial)
from sklearn.preprocessing import LabelEncoder
cities=['kathmandu','pokhara','butwal']
le=LabelEncoder()
encoded = le.fit_transform(cities)
encoded
```

```python
# one hot encoding
df = pd.DataFrame({'city':['kathmandu','pokhara','butwal']})
encoded = pd.get_dummies(df, columns=['city'])
encoded
```

```python
from sklearn.preprocessing import OrdinalEncoder
quality = [['poor'],['medium'],['high']]
encoder = OrdinalEncoder()
encoded = encoder.fit_transform(quality)
encoded
```
