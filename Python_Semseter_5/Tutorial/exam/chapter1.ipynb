```python
# for loop under the hood using iterator
n=[10,20,30,40,50]
it =iter(n)

while True:
    try:
        x=next(it)
        print(x)
    except StopIteration:
        break
```

```python
def squares(n):
    for i in range(n):
        yield i*i
    
for x in squares(5):
    print(x)
```

```python
# decorator
from datetime import datetime 
def mydecorator(func):
    def wrapper(*args, **kwargs):
        print("before")
        print(f"function name: {func.__name__}")
        start = datetime.now()
        print(start)
        result = func(*args, **kwargs)
        print("after")
        return result
    return wrapper

@mydecorator
def functionthis():
    print("hello")

functionthis()
```

```python
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def __str__(self):
        return f"{self.real} + {self.img} i"
    
    def __add__(self, other):
        r = self.real + other.real
        i = self.img + other.img
        return Complex(r,i)
    
a=Complex(3, 4)
b = Complex(2, 5)

print(a)
print(b)

c = a + b
print(c)
        
```

```python
class Animal:
    def __init__(self,name):
       self.name = name

    def speak(self):
        print(f"{self.name} is speaking")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def speak(self):
        super().speak()
        print(f"{self.breed} is a dog")

d1 = Dog('tommy', 'bulldog')
d1.speak() 
```

```python
# logging
import logging

logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s %(levelname)s %(message)s",
    filename="logs.log",
    filemode="a"
    )
logging.info("program started")
logging.warning("this is warning")
logging.error("This is error")

with open("logs.log",'r') as f:
    rows = f.readlines()
    for row in rows:
        print(row)
```
