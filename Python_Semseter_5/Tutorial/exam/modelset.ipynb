```python
import csv
import json

def csv_to_json(csv_filename, json_filename):
    try:
        students = []
        with open(csv_filename, 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            
            for row in reader:
                try:
                    # Example: ensure "age" is an integer
                    row["age"] = int(row["age"])
                    students.append(row)
                except ValueError:
                    print(f"Skipping malformed row: {row}")
        
        # Export to JSON
        with open(json_filename, 'w') as jsonfile:
            json.dump(students, jsonfile, indent=4)
        print(f"Data exported successfully to {json_filename}")

    except FileNotFoundError:
        print(f"Error: The file '{csv_filename}' was not found.")

    except csv.Error as e:
        print(f"Error reading CSV file: {e}")

    finally:
        print("Process completed.")


# Example usage
csv_to_json("students.csv", "students.json")
```

```python
class Complex:
    def __init__(self, real, img):
        self.real = real 
        self.img  = img

    def __str__(self):
        return f"{self.real} + {self.img } i"

    def __add__(self, other):
        r = self.real + other.real
        i = self.img + other.img
        return Complex(r,i)

c1 = Complex(2,4)
c2 = Complex(5,1)

c3 = c1 + c2
print(c3)
```

```python
def read_lines(filename):
    with open(filename) as f:
        for line in f:
            yield line

for line in read_lines("practice.txt"):
    print(line)
```
