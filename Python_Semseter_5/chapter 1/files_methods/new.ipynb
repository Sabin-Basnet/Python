```python
# count frequency of words
with open(file='try.txt' , mode='r') as f:
    content = f.read().lower()
    dict={}
    words=content.split()
    for word in words:
        if word in dict:
            dict[word] +=1
        else:
            dict[word]=1

    print(dict)
    
```

```python
# write in files 
with open(file='new.txt' , mode='w') as f:
    while(True):
        intput_text= input("Enter text(press Q or q to quit): ")
        if(intput_text == 'Q' or intput_text == 'q'):
            break
        f.writelines(intput_text)
        f.write("\n")
        
```

```python
# using dicitionary as input in csv file
import csv
data=[
    {'Name': 'John', 'Address': 'Biratnagar', 'Age': '23'},
    {'Name': 'Hari', 'Address': 'Dharan', 'Age': '32'},
    {'Name': 'Ram', 'Address': 'Sunsari', 'Age': '14'}
]

with open(file='file1.csv' , mode='w', newline="") as f:
    writer=csv.DictWriter(f, fieldnames=["Name",'Address','Age'])
    writer.writeheader()
    writer.writerows(data)
```
