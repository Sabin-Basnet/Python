```python

add= lambda a,b:a+b

add(4,2)
# use lambda when you need short function. Lambda is nameless function.
```

```python
# use case-1: sorting
data=[(1,"Ram","E"),(3,"Shyam","W"),(2,"Hari","A")]
sorted_data= sorted(data,key=lambda x: x[2])
print(sorted_data)
```

```python
# use case-2 : filtering
aList= [1,2,3,4]
evens = list(filter(lambda x:x%2==0, aList))
print(evens)
```

```python
# use case-3: mapping
aList=[1,2,3,4]
squares= list(map(lambda x:x**2, aList))
print(squares)
```
