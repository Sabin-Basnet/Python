```python
from collections import Counter

nums=[1,2,2,3,3,3]
c= Counter(nums)
print(type(c))
```

```python
from collections import defaultdict
aDict = defaultdict(int)
aDict["a"]=aDict["a"] + 1
print(aDict)
```

```python
from collections import deque
q=deque([1,2,3])
q.popleft()
print(q)
```

```python
from collections import namedtuple
Point = namedtuple('Point','x y')
p = Point(3, 4)
print(p)
```
