```python
aList = [1,2,3,4]
it = iter(aList)

while True:
    try:
        n=next(it)
        print(n)

    except StopIteration:
        break
```

```python
def squares(x):
    for i in range(x):
        yield i*i

# print(squares(5))
# print(type(squares(5)))

for i in squares(5):
    print(i)
```

```python
def decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper
```

```python
@decorator
def sayhello():
    print("Hello")

sayhello()
```

```python
import time 
def timers(func):
    def wrapper(*args,**kwargs):
        start=time.time()
        result=func(*args,**kwargs)
        end= time.time()
        print(f"It took : {end-start} seconds")
        return result
    return wrapper()
```

```python
@timers
def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*i
        n=n+1
    return f

fact(100)
```
