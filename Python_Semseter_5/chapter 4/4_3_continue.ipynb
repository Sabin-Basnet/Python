```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
```

```python
# y= m*X + c
X = np.arange(0,30,0.277)
y= 0.9912 * X + 0.7734
print(len(X),len(y))
```

```python
# split
split = int(len(X)*0.8)

X_train , y_train = X[:split], y[:split]
X_test, y_test = X[split:], y[split:]
```

```python
def plot_data(X_train = X_train, y_train = y_train,
              X_test=X_test, y_test=y_test, predictions=None):
    plt.scatter(X_train, y_train, color= 'black', label="Training data")
    plt.scatter(X_test, y_test, color= 'red', label="Testing data")
    if predictions is not None:
        plt.scatter(X_test, predictions, color= 'green', label="Predicted data")

    plt.title('X vs y')
    plt.xlabel("X")
    plt.ylabel('y')
    plt.legend()
    plt.show()
```

```python
plot_data(X_train, y_train)
```

```python
# add constant for statsmodel explictilly
X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

# summary
model = sm.OLS(y_train, X_train_sm).fit()
print(model.summary())
```

```python
# multiple regression data
X1 = np.arange(0,30,0.277)
X2 = np.arange(0,30,0.277)

y= 0.9912 * X1 +  2.3 *X2 + 0.7734

df = pd.DataFrame({
    'X1':X1, 'X2':X2 ,'y':y
})

X= df.drop(columns='y')
y=df['y']
print(len(X), len(y))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2 , random_state=7)
```

```python
# add constant for statsmodel explictilly
X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

# summary
model = sm.OLS(y_train, X_train_sm).fit()
print(model.summary())
```
