```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y= m*X + c
X = np.arange(0,30,0.277)
y= 0.9912 * X + 0.7734
print(len(X),len(y))
```

```python
X_mean = np.mean(X)
y_mean = np.mean(y)
print(X_mean)
print(y_mean)
```

```python
n = len(X)
cov_x_y = 1/n * np.sum((X-X_mean)*(y-y_mean))
print(cov_x_y)
cov_x = 1/n * np.sum(np.square(X-X_mean))
print(cov_x)
```

```python
b1 = cov_x_y / cov_x
print(b1)
```

```python
b0 = y_mean - b1*X_mean
print(f"y= {b0:.4f} + {b1:.4f}*X")
```

Doing it with library

```python
from sklearn.linear_model import LinearRegression

X = np.arange(0,30,0.277)
y= 0.9912 * X + 0.7734
print(len(X),len(y))
```

```python
split = int(len(X)*0.8)

X_train , y_train = X[:split], y[:split]
X_test, y_test = X[split:], y[split:]
```

```python
def plot_data(X_train = X_train, y_train = y_train,
              X_test=X_test, y_test=y_test, predictions=None):
    plt.scatter(X_train, y_train, color= 'black', label="Training data")
    plt.scatter(X_test, y_test, color= 'red', label="Testing data")
    if predictions is not None:
        plt.scatter(X_test, predictions, color= 'green', label="Predicted data")

    plt.title('X vs y')
    plt.xlabel("X")
    plt.ylabel('y')
    plt.legend()
    plt.show()
```

```python
model = LinearRegression()
model.fit(X_train.reshape(-1,1), y_train.reshape(-1,1))

predictions = model.predict(X_test.reshape(-1,1))
```

```python
plot_data(X_train, y_train,X_test, y_test,predictions)
```
