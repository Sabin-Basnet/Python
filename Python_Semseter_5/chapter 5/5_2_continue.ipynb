```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(np.random.standard_normal(50).cumsum(), color='black', linestyle='dashed')
ax1.plot(np.random.standard_normal(50).cumsum(), color='black', linestyle='dashed')
ax2.hist(np.random.standard_normal(100),bins=10, edgecolor='black')
ax3.scatter(np.arange(30),np.arange(30)+2*np.random.standard_normal(30),edgecolor='black')

ax4.bar(x=['male','female'],height=[2,3])

fig.show()
```

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey= True)
print(fig)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.standard_normal(500),bins=50, edgecolor=
                       'black', alpha=0.7)
        
fig.subplots_adjust(wspace=0.1, hspace=.1)
```

```python
x= np.arange(1,6)
y=x*2
data = np.random.randn(100)
```

```python
# create subplots
fig, axes = plt.subplots(2,2, figsize=(6,4))

# line plot
axes[0,0].plot(x,y)
axes[0,0].set_title("Line plot")

# bar chart
axes[0,1].bar(x,y, edgecolor='black')
axes[0,1].set_title("bar chart")

# histogram
axes[1,0].hist(data, edgecolor='black')
axes[1,0].set_title("histogram")

# scatter plot
axes[1,1].scatter(x,y)
axes[1,1].set_title("Scatter plot")

# fix spacing
plt.tight_layout()

```
