```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(1,50,0.5)
y=np.square(X)
```

```python
plt.plot(X,y, color='r', alpha=0.3, marker=1)
plt.xlabel("X")
plt.ylabel("X - squared")
plt.show()
```

```python
categories = ['Apple','Banana','Mango']
count =[20,12,27]
color=['red','green','blue']
plt.bar(categories,count, width= 0.6, edgecolor= 'red', color=['green','gray','blue'])
plt.xticks(rotation=90)
# plt.legend()
plt.show()
```

```python
data=np.random.normal(loc = 50, scale=10, size=1000)

plt.hist(data, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(
    n_features=2, n_samples=1000, random_state=7
)
print(X.shape)
len(y)
```

```python
plt.scatter(X[:, 0],X[:, 1], c=y, cmap='viridis', edgecolors='black')
plt.colorbar(label='Cluster')
plt.show()
```
