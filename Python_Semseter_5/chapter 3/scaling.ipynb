```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=pd.Series([0,1,10,100,1000])
x_log=np.log1p(x)
print(x_log)
```

```python
plt.plot(x,label='x')
plt.plot(x_log,label='x_log')
plt.legend()
```
