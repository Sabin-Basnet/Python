```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'ID':[1,2,3],
    'Name':['Ram','Shyam','Hari'],
    'Level':['low','medium','high']
})
```

```python
df["Level"]=df['Level'].astype('category')
```

```python
df['Level'].cat.codes
print(df)
```

```python
encoded = pd.get_dummies(df['Level'],prefix='Lev',dtype=int)
print(encoded)
```

```python
df_final= pd.concat([df,encoded],axis=1)
print(df_final)
```

```python
df = pd.DataFrame({
    'Date':['2026-01-01','2026-01-02','2026-01-03','2026-01-04'],
    'Sales': [200,302,201,150]
})
df.info()
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df=df.set_index('Date')
df
```

```python
df['Year']=df.index.year
df['Month']=df.index.month
df['Day']=df.index.day_name()
df
```

```python
df.loc['2026-01-01':'2026-01-02']
```

```python
df.loc['2026-01']
```

```python
df.to_csv('dates.csv')
```
