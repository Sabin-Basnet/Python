```python
import random
import pandas as pd
import tqdm
```

```python
names=['Ram','Shyam','Hari','Sita','Gopal']
final_list=[]
for i in tqdm.tqdm(range(1,10000)):
    random_name = names[random.randint(0,4)]
    final_list.append({
        'Eid':i,
        'Name':random_name ,
        'Salary': 100000*random.random()
    })
    df=pd.DataFrame(final_list)
    df.to_csv('big_data.csv')
```

```python
first_chunk= True
output_file='final_big_data.csv'
for chunk in pd.read_csv('big_data.csv', chunksize=10000):
    chunk['Gender']=chunk['Name'].apply(lambda x: 'M' if x in ['Ram','Shyam','Hari','Gopal'] else 'F')
    chunk.to_csv(
        output_file,
        header=first_chunk,
        index=False,
        mode='w' if first_chunk else 'a'
    )
    first_chunk=False
```
