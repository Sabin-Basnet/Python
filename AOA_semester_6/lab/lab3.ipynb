```python
def quickSort(A, low, high): 
    if low < high:
        p = partition(A, low, high)
        quickSort(A, low, p - 1)
        quickSort(A, p + 1, high)

def partition(A, low, high):
    pivot = A[low]
    x = low + 1
    y = high

    while x <= y:
    # Move x right while elements are <= pivot 
        while x <= high and A[x] <= pivot:
            x += 1
    # Move y left while elements are > pivot 
        while y >= low + 1 and A[y] > pivot:
            y -= 1
    # If pointers haven't crossed, swap 
        if x < y:
            A[x], A[y] = A[y], A[x]
    A[low], A[y] = A[y], A[low]
    return y


arr = [34, 72, 33, 77, 15, 55, 86]
print("Original array:", arr) 
quickSort(arr, 0, len(arr) - 1)
print("Sorted array:", arr)
```

```python
def heapSort(A):
    n = len(A)
    # Build a max heap
    buildMaxHeap(A, n)
    
    # One by one extract elements from the heap
    for i in range(n - 1, 0, -1):
        # Move current root to end
        A[0], A[i] = A[i], A[0]
        # Call maxHeapify on the reduced heap
        maxHeapify(A, 0, i)

def buildMaxHeap(A, n):
    # Index of the last non-leaf node
    start = (n // 2) - 1
    # Perform heapify on each non-leaf node from bottom up
    for i in range(start, -1, -1):
        maxHeapify(A, i, n)

def maxHeapify(A, i, n):
    L = 2 * i + 1
    R = 2 * i + 2
    largest = i

    # If left child is larger than root
    if L < n and A[L] > A[largest]:
        largest = L

    # If right child is larger than largest so far
    if R < n and A[R] > A[largest]:
        largest = R

    # If largest is not root, swap and continue heapifying
    if largest != i:
        A[i], A[largest] = A[largest], A[i]
        maxHeapify(A, largest, n)


arr = [34, 72, 33, 77, 15, 55, 86]
print("Original array:", arr) 
heapSort(arr)
print("Sorted array:", arr)
```
