```python
def feasible_sub_set(i, current_sum, M, remSum):
    if current_sum > M or (current_sum + remSum) < M:
        return 0  # not feasible
    else:
        return 1  # feasible

def sub_set_problem(i, current_sum, M, remSum, w, X):
    if feasible_sub_set(i, current_sum, M, remSum) == 0:
        return
        
    if current_sum == M:
        print("Solution found (X vector):", X[1:i+1])
        return

    # Check if there are still elements left to process
    if i + 1 < len(w):
        # Include next item (i+1)
        X[i+1] = 1
        sub_set_problem(i+1, current_sum + w[i+1], M, remSum - w[i+1], w, X)
        
        # Exclude next item
        X[i+1] = 0
        sub_set_problem(i+1, current_sum, M, remSum - w[i+1], w, X)

# Elements weights (Using 0 at index 0 to make it 1-indexed)
w = [0, 5, 10, 15, 20] 

# Solution array initialized to 0 (length matches w)
X = [0] * len(w)

total_sum = sum(w)

print(f"Given weights: {w[1:]} | Target Sum: {M}\n")
sub_set_problem(i=0, current_sum=0, M=M, remSum=total_sum, w=w, X=X)
```
