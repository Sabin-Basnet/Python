Practical 1: Implementation and Complexity Analysis of Iterative, Numeric,
and Recursive Algorithms in Python

```python
# a. Iterative Algorithm– Fibonacci (Iterative Approach)
def fibo(n):
    if n==1:
        return 0
    if n==2:
        return 1
    
    a,b = 0, 1
    for i in range(3, n+1): 
        c=a+b
        a=b
        b=c
    return c

# time complexity : O(n)

n=8
print(f"{n}th number of fibo series = ",fibo(n))
```

```python
# b. Recursive Algorithm– Fibonacci (Recursive Approach)
def fibonacci_recursive(n):
    if n==1:
        return 0
    if n==2:
        return 1
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

# time complexity : O(2^n)
# space complexity : 
n=8
print(f"{n}th number of fibo series = ",fibonacci_recursive(n))
```

```python
# c. NumericAlgorithm–EuclideanAlgorithm(GCD)
def gcd_algorithm(a:int, b:int):
    if a==0 and b==0:
        return "Undefined GCD"
    if b==0:
        return a
    if a==0:
        return b
    while b!=0:
        remainder=a%b
        a=b
        b=remainder
    return a

# time complexity : O( log(min(a,b)) )
# space complexity : O(1)
a=12
b=4
print(f"gcd of {a} and {b} is {gcd_algorithm(a, b)}")
```

```python
# d. RecursiveAlgorithm–EuclideanAlgorithm(GCD)
def gcd_recursive ( a, b):
    if a==0 and b==0 :
        return "Undefined GCD"
    if b==0:
        return a
    return gcd_recursive(b, a%b)
    
# time complexity : O( log(min(a,b) )
# space complexity : O( log(min(a,b) )
a=12
b=4
print(f"gcd of {a} and {b} is {gcd_recursive(a, b)}")
```

```python
# e. Sequential Search (Iterative)
def sequential_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
    
arr = [3, 4, 7, 1, 3, 9, 0]
key = 9
a = sequential_search(arr, key)
# print(a)
if( a != -1):
    print(f"Yes, the key is found at position : {a+1} th")
else:
    print("Key Not Found")

# time complexity: O(n)
# space complexity: O(1)
```
