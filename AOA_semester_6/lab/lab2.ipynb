```python
def greedy_knapsack(X, V, W, M):
    # X = array of items
    # V = array of profits of each item
    # W = array of weight of each item
    # M = capacity of knapsack
    n = len(X)
    items = []
    
    # 1. Calculate ratios and pack item data
    for i in range(n):
        ratio = V[i] / W[i]
        items.append((ratio, X[i], V[i], W[i]))

    # 2. Sort outside the loop (Fixed Indentation)
    items.sort(reverse=True)
    
    s = []    # selected items
    sw = 0    # weight of selected items
    sp = 0    # profit of selected items
    i = 0
    
    # 3. Process items greedily (Fixed Indentation)
    while i < n:
        ratio, x, p, w = items[i]
        if (sw + w) <= M:
            s.append(x)
            sw = sw + w
            sp = sp + p
        else:
            frac = (M - sw) / w
            s.append((x, round(frac, 2)))  # Added rounding for clean display
            sp = sp + p * frac
            sw = sw + w * frac
            break
        i += 1
        
    return s, sw, sp
    
M = 20
X = ['x1', 'x2', 'x3']
V = [24, 25, 15]
W = [12, 13, 14]

s, sw, sp = greedy_knapsack(X, V, W, M)
print(f"s = {s}\nsw = {sw}\nsp = {sp}")
```
