```python
def is_safe(board, x,y,n):
    # column test
    for row in range(x):
        if (board[row][y] == 'Q'):
            return False
    
    # left upper diagonal test
    r = x-1
    c = y-1
    while( r>=0 and c>=0):
        if (board[r][c] == 'Q'):
            return False 
        r-=1
        c-=1

    # right upper diagonal test
    r=x-1
    c=y+1
    while( r>=0 and c<n):
        if (board[r][c] == 'Q'):
            return False 
        r-=1
        c+=1
    
    return True
```

```python
from pprint import pprint
n=4
board = [['_']*n for i in range(n)]
board[0][0] = 'Q'
pprint(board)
print(is_safe(board=board,x= 0,y=0,n=n))
```

```python
n = int(input("Enter number of queens: "))
board = [['_']*n for i in range(n)]
def nQueenSolver(board, row, n):
    if(row>=n):
        return True
    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = 'Q'
            if(nQueenSolver(board=board, row= row+1, n=n)):
                return True
            board[row][col] = '_'
    return False

if(nQueenSolver(board, 0, n)):
    pprint(board)
else:
    print("No solution")
```

```python
# Task 1: Goat, Wolf and Cabbage Problem
from collections import deque

F, W, G, C = 0, 1, 2, 3

def is_valid(state):
    """Checks if a state violates any constraints."""
    farmer, wolf, goat, cabbage = state
    
    # If wolf and goat are together but farmer is on the other side
    if wolf == goat and farmer != wolf:
        return False
    # If goat and cabbage are together but farmer is on the other side
    if goat == cabbage and farmer != goat:
        return False
        
    return True

def get_successors(state):
    """Generates all valid next states from the current state."""
    successors = []
    farmer_side = state[F]
    next_side = 'E' if farmer_side == 'W' else 'W'
    
    # Action 1: Farmer crosses alone
    next_state = list(state)
    next_state[F] = next_side
    if is_valid(next_state):
        successors.append((tuple(next_state), "Farmer crosses alone"))
        
    # Actions 2-4: Farmer takes one item
    items = [("wolf", W), ("goat", G), ("cabbage", C)]
    for item_name, index in items:
        if state[index] == farmer_side:  # Item is on the same side as the farmer
            next_state = list(state)
            next_state[F] = next_side    # Move farmer
            next_state[index] = next_side # Move item
            if is_valid(next_state):
                successors.append((tuple(next_state), f"Farmer takes {item_name}"))
                
    return successors

def solve_puzzle():
    initial_state = ('W', 'W', 'W', 'W')
    goal_state = ('E', 'E', 'E', 'E')
    
    # Queue stores tuples of (current_state, path_taken)
    queue = deque([(initial_state, [])])
    visited = {initial_state}
    
    while queue:
        current_state, path = queue.popleft()
        
        # Check if goal is reached
        if current_state == goal_state:
            return path
            
        for next_state, action in get_successors(current_state):
            if next_state not in visited:
                visited.add(next_state)
                queue.append((next_state, path + [action]))
                
    return None

# execution
if __name__ == "__main__":
    solution = solve_puzzle()
    if solution:
        print("Sequence of moves to solve the puzzle:")
        for step, action in enumerate(solution, 1):
            print(f"Step {step}: {action}")
    else:
        print("No solution found.")
```

```python
# Task 2: Water Jug Problem
from collections import deque

def solve_water_jug():
    # size of the jugs iin liters
    max_a, max_b = 4, 3
    
    # BFS Queue stores tuples of: (current_state, path_taken)
    # Initial State: (0, 0)
    queue = deque([((0, 0), [(0, 0)])])
    
    # Visited set to prevent infinite loops
    visited = set([(0, 0)])
    
    while queue:
        (a, b), path = queue.popleft()
        
        # Goal Check: Jug A has exactly 2 liters
        if a == 2:
            return path
            
        # Generate all possible next states
        possible_moves = [
            (max_a, b),   # Fill Jug A
            (a, max_b),   # Fill Jug B
            (0, b),       # Empty Jug A
            (a, 0),       # Empty Jug B
            # Pour A -> B (pour until B is full or A is empty)
            (max(0, a - (max_b - b)), min(max_b, b + a)),
            # Pour B -> A (pour until A is full or B is empty)
            (min(max_a, a + b), max(0, b - (max_a - a)))
        ]
        
        for next_state in possible_moves:
            if next_state not in visited:
                visited.add(next_state)
                # Append the new state to the path history
                queue.append((next_state, path + [next_state]))
                
    return None

# Execute the solver and print the path
solution_path = solve_water_jug()

if solution_path:
    print("Shortest path found to solve the Water Jug Problem:")
    for step, state in enumerate(solution_path):
        print(f"Step {step}: Jug A = {state[0]}L, Jug B = {state[1]}L")
else:
    print("No solution could be found.")
```

```python
# Task 3: Robot Navigation Problem
from collections import deque

def solve_robot_navigation():
    # 5x5 Grid representation: 0 = Open path, 1 = Obstacle (X)
    # S is at (0,0), G is at (4,4)
    grid = [
        [0, 0, 0, 0, 0],  # Row 0
        [0, 1, 1, 0, 0],  # Row 1
        [0, 0, 0, 1, 0],  # Row 2
        [1, 1, 0, 0, 1],  # Row 3
        [0, 0, 0, 0, 0]   # Row 4
    ]
    
    rows, cols = 5, 5
    start = (0, 0)
    goal = (4, 4)
    
    # BFS Queue stores tuples of: (current_position, path_taken)
    queue = deque([(start, [start])])
    visited = set([start])
    
    # Define directional actions: (row_offset, col_offset)
    actions = {
        "Up": (-1, 0),
        "Down": (1, 0),
        "Left": (0, -1),
        "Right": (0, 1)
    }
    
    while queue:
        (r, c), path = queue.popleft()
        
        # Goal Check
        if (r, c) == goal:
            return path
            
        # Explore valid neighboring actions
        for action, (dr, dc) in actions.items():
            nr, nc = r + dr, c + dc
            
            # Check bounds and if the cell is unvisited and free of obstacles
            if 0 <= nr < rows and 0 <= nc < cols:
                if grid[nr][nc] == 0 and (nr, nc) not in visited:
                    visited.add((nr, nc))
                    queue.append(((nr, nc), path + [(nr, nc)]))
                    
    return None

# Execute solver and print results
shortest_path = solve_robot_navigation()

if shortest_path:
    print("Shortest path found by BFS:")
    for step, coords in enumerate(shortest_path):
        print(f"Step {step}: {coords}")
else:
    print("No valid path exists to the goal.")
```
